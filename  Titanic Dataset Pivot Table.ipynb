```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
d=pd.read_csv('/home/soniya/Documents/train.csv')
```

```python
d
```

```python
d.head()
```

```python
d.drop(['PassengerId','Ticket','Name'],inplace=True,axis=1)
```

```python
d
```

```python
#a single index
table=pd.pivot_table(data=d,index=['Sex'])
```

```python
table
```

```python
table.plot(kind='bar');
```

```python
#multiple indexes
table=pd.pivot_table(d,index=['Sex','Pclass'])
```

```python
table
```

```python
#multiple indexes
table = pd.pivot_table(d,index=['Sex','Pclass'])
table
```

```python
#different aggregate functions
table = pd.pivot_table(d,index=['Sex','Pclass'],aggfunc={'Age':np.mean,'Survived':np.sum})
table
```

```python
table = pd.pivot_table(d,index=['Sex','Pclass'],values=['Survived'], aggfunc=np.mean)
table
```

```python
table.plot(kind='bar');

```

```python
#columns
table = pd.pivot_table(d,index=['Sex'],columns=['Pclass'],values=['Survived'],aggfunc=np.sum)
table
```

```python
table.plot(kind='bar');
```

```python
#display null values
table = pd.pivot_table(d,index=['Sex','Survived','Pclass'],columns=['Embarked'],values=['Age'],aggfunc=np.mean)
table
```

```python
#handling null values
table = pd.pivot_table(d,index=['Sex','Survived','Pclass'],columns=['Embarked'],values=['Age'],aggfunc=np.mean,fill_value=np.mean(d['Age']))
```
